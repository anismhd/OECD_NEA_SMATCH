```python
with open('EPR_mat.pickle', 'rb') as handle:
    EPR_mat = pickle.load(handle)
with open('GROUPS.pickle', 'rb') as handle:
    GROUPS = pickle.load(handle)
with open('ELEM_GROUPS.pickle', 'rb') as handle:
    ELEM_GROUPS = pickle.load(handle)
with open('NODE_GROUPS.pickle', 'rb') as handle:
    NODE_GROUPS = pickle.load(handle)
```

```python
EPR_geo = UNV('InputsV2/PHASE_3_STAGE_1_INPUT_v2/02_Model/CRUAS_NPP.unv')
```

```python
FreeNodes = np.loadtxt('InputsV2/PHASE_3_STAGE_1_INPUT_v2/02_Model/FreeNodes.dat', dtype=int, delimiter=',')
```

# Lumped Mass

```python
Lumped_Mass_Groups = {}
Lumped_Mass = {}
for gid, details in EPR_mat.section_details['M_T_N'].items():
    node_set = []
    for eid in ELEM_GROUPS[gid]:
        if EPR_geo.fem['elements'][161][eid][0] in FreeNodes:
            continue
        node_set.append(EPR_geo.fem['elements'][161][eid][0])
    Lumped_Mass_Groups[gid] = node_set
    if not(node_set):
        continue
    Lumped_Mass[gid] = {'nSet':node_set, 'mass': [details['parameters'][0],details['parameters'][1],details['parameters'][2],0,0,0]}
for gid, details in EPR_mat.section_details['M_TR_D_N'].items():
    node_set = []
    for eid in ELEM_GROUPS[gid]:
        if EPR_geo.fem['elements'][161][eid][0] in FreeNodes:
            continue
        node_set.append(EPR_geo.fem['elements'][161][eid][0])
    Lumped_Mass_Groups[gid] = node_set
    if not(node_set):
        continue
    if len(details['parameters']) == 3:
        Lumped_Mass[gid] = {'nSet':node_set, 'mass': [details['parameters'][0],details['parameters'][0],details['parameters'][0],\
                            details['parameters'][1],details['parameters'][2],0]}
    else:
        Lumped_Mass[gid] = {'nSet':node_set, 'mass': [details['parameters'][0],details['parameters'][0],details['parameters'][0],\
                            details['parameters'][1],details['parameters'][2],details['parameters'][3]]}
```

```python
with open('Lumped_Mass_Groups.pickle', 'wb') as handle:
    pickle.dump(Lumped_Mass_Groups, handle, protocol=2)
with open('Lumped_Mass.pickle', 'wb') as handle:
    pickle.dump(Lumped_Mass, handle, protocol=2)
```
